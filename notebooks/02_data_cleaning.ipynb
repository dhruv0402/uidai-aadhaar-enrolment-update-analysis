```python
import pandas as pd

ENROL_URL = "https://raw.githubusercontent.com/dhruv0402/uidai-aadhaar-enrolment-update-analysis/refs/heads/main/data/raw/aadhaar_enrolment_master.csv"
DEMO_URL  = "https://raw.githubusercontent.com/dhruv0402/uidai-aadhaar-enrolment-update-analysis/refs/heads/main/data/raw/aadhaar_demographic_master.csv"
BIO_URL   = "https://raw.githubusercontent.com/dhruv0402/uidai-aadhaar-enrolment-update-analysis/refs/heads/main/data/raw/aadhaar_biometric_master.csv"

enrol = pd.read_csv(ENROL_URL)
demo  = pd.read_csv(DEMO_URL)
bio   = pd.read_csv(BIO_URL)

print(enrol.shape, demo.shape, bio.shape)
enrol.head()
import os

REPO_NAME = "uidai-aadhaar-enrolment-update-analysis"   # ðŸ‘ˆ change this
PROJECT_ROOT = os.path.join(os.getcwd(), REPO_NAME)

print("âœ… Current working directory:", os.getcwd())
print("âœ… Project root:", PROJECT_ROOT)

# Move into notebooks folder if you have it (optional)
# os.chdir(os.path.join(PROJECT_ROOT, "notebooks"))

# We'll use a base path approach instead of chdir ( safer)
!mkdir -p /content/uidai-aadhaar-enrolment-update-analysis/data/raw
!mkdir -p /content/uidai-aadhaar-enrolment-update-analysis/data/processed
print("âœ… raw and processed folders created")

!rm -rf uidai-aadhaar-enrolment-update-analysis
print("âœ… Old folder removed (if it existed)")
!git clone https://github.com/dhruv0402/uidai-aadhaar-enrolment-update-analysis.git
%cd /content/uidai-aadhaar-enrolment-update-analysis



def inspect_df(df, name):
    print(f"\n==================== {name.upper()} ====================")
    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())
    print("\nData Types:")
    print(df.dtypes)
    print("\nMissing Values (Top 15):")
    print(df.isna().sum().sort_values(ascending=False).head(15))
    display(df.head(5))

inspect_df(enrol, "enrolment")
inspect_df(demo, "demographic")
inspect_df(bio, "biometric")
import numpy as np
import pandas as pd

KEY_COLS = ["date", "state", "district", "pincode"]

def clean_column_names(df):
    df = df.copy()
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(r"\s+", "_", regex=True)
    )
    return df

def normalize_text(series):
    s = series.astype(str).str.strip()
    s = s.str.replace(r"\s+", " ", regex=True)
    s = s.str.title()
    s = s.replace({"Nan": np.nan, "None": np.nan, "Null": np.nan, "": np.nan})
    return s

def normalize_district(series):
    s = series.astype(str).str.strip()
    s = s.str.replace(r"\s+", " ", regex=True)
    s = s.str.replace(r"[-_/]+", " ", regex=True)
    s = s.str.replace(r"\s+", " ", regex=True)
    s = s.str.title()
    s = s.replace({"Nan": np.nan, "None": np.nan, "Null": np.nan, "": np.nan})
    return s

def normalize_pincode(series):
    s = series.astype(str).str.strip()
    s = s.replace({"nan": np.nan, "None": np.nan, "null": np.nan, "": np.nan})
    s = s.str.replace(r"\\D", "", regex=True)
    s = s.where(s.str.len() == 6, np.nan)
    return s
```

```python
def clean_uidai_dataset(df, name):
    df = df.copy()

    # Column names
    df = clean_column_names(df)

    # Required columns check
    missing = [c for c in KEY_COLS if c not in df.columns]
    if missing:
        raise ValueError(f"{name}: Missing required columns: {missing}")

    # Date parsing
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df = df.dropna(subset=["date"])

    # Normalize text fields
    df["state"] = normalize_text(df["state"])
    df["district"] = normalize_district(df["district"])
    df["pincode"] = normalize_pincode(df["pincode"])

    # Drop if state/district missing
    df = df.dropna(subset=["state", "district"])

    # Numeric conversion
    count_cols = [c for c in df.columns if c not in KEY_COLS]
    for c in count_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    df[count_cols] = df[count_cols].fillna(0)

    # Remove duplicates safely
    df = df.drop_duplicates()
    df = df.groupby(KEY_COLS, as_index=False)[count_cols].sum()

    return df
```

```python
enrol_clean = clean_uidai_dataset(enrol, "enrolment")
demo_clean  = clean_uidai_dataset(demo, "demographic")
bio_clean   = clean_uidai_dataset(bio, "biometric")

print("Cleaned shapes:")
print("Enrolment :", enrol_clean.shape)
print("Demographic:", demo_clean.shape)
print("Biometric :", bio_clean.shape)
```

```python
GROUP_KEYS = ["state", "district", "month"]

def aggregate_district_month(df):
    df = df.copy()

    # Month extraction (YYYY-MM)
    df["month"] = df["date"].dt.to_period("M").astype(str)

    exclude = {"date", "pincode", "month", "state", "district"}
    value_cols = [c for c in df.columns if c not in exclude]

    out = (
        df
        .groupby(GROUP_KEYS, as_index=False)[value_cols]
        .sum()
    )

    return out

enrol_dm = aggregate_district_month(enrol_clean)
demo_dm  = aggregate_district_month(demo_clean)
bio_dm   = aggregate_district_month(bio_clean)
```

```python
for df in [enrol_dm, demo_dm, bio_dm]:
    df.drop(columns=["dataset"], errors="ignore", inplace=True)
```

```python
# Force string type
enrol_dm["state"] = enrol_dm["state"].astype(str)
enrol_dm["district"] = enrol_dm["district"].astype(str)

# Remove fake UIDAI placeholder rows
enrol_df_month = enrol_dm[
    (enrol_dm["state"] != "100000") &
    (enrol_dm["district"] != "100000")
].reset_index(drop=True)
```

```python
os.makedirs("data/processed", exist_ok=True)

enrol_df_month.to_csv("../data/processed/enrol_df_month.csv", index=False)
demo_dm.to_csv("../data/processed/demo_dm.csv", index=False)
bio_dm.to_csv("../data/processed/bio_dm.csv", index=False)

print("âœ… Final processed datasets written:")
print(" - enrol_df_month.csv")
print(" - demo_dm.csv")
print(" - bio_dm.csv")
```
